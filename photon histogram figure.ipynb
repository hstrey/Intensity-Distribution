```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, poisson

SMALL_SIZE = 16
MEDIUM_SIZE = 18
BIGGER_SIZE = 20

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

SEED = 350107321 # from random.org
np.random.seed(SEED)

print(plt.style.available)
plt.style.use('seaborn-white')
```

```python
xlim = 3
x = np.linspace(-xlim,xlim)
n = norm.pdf(x)
xb = np.array([-xlim,-np.sqrt(np.pi/2)/2-0.0000001,-np.sqrt(np.pi/2)/2,np.sqrt(np.pi/2)/2,np.sqrt(np.pi/2)/2+0.0000001,xlim])
pb = np.array([0,0,1/np.sqrt(np.pi/2),1/np.sqrt(np.pi/2),0,0])
```

```python
# this is how many particles are in the box of there is on average one particle in the box
print(6/np.sqrt(np.pi/2))
```

```python
xf = np.random.random(5)*2*xlim-xlim
yfg = norm.pdf(xf)
inorout = (-np.sqrt(np.pi/2)/2<xf)&(xf<np.sqrt(np.pi/2)/2)
yfb = np.where(inorout,np.ones(5)/np.sqrt(np.pi/2),np.zeros(5))
print(xf)
```

```python
plt.plot(x,n,label="Gaussian")
plt.plot(xb,pb,label="Box")
plt.scatter(xf,yfg)
plt.scatter(xf,yfb)

plt.xlabel("x")
plt.ylabel("p(x)")
plt.savefig("Gaussian_vs_box.png",format="png",dpi=300,bbox_inches='tight',facecolor="white",backgroundcolor="white")
```

```python
xp = np.arange(5)
pp = poisson(mu=1).pmf(xp)
fig, ax = plt.subplots(1, 1)
ax.vlines(xp, 0, pp, colors='k', linestyles='-', lw=3,label=r'Poisson $\mu=1$')
plt.legend()
plt.ylabel("p(I)")
plt.xlabel("I")
plt.savefig("Poisson.png",format="png",dpi=300,bbox_inches='tight',facecolor="white",backgroundcolor="white")
```

```python
xlim = 2
x = np.linspace(-xlim,xlim,2000)
n = norm.pdf(x)
n[0]=0
n[-1]=0
xf = np.random.random(1)*2*xlim-xlim
yfg = norm.pdf(xf)
print(xf)

plt.plot(x,n)
plt.scatter(xf,yfg)
plt.xlabel("x")
plt.ylabel(r"$\Phi(x)$")
plt.ylim((0,0.45))
plt.xticks([-2,2],["-L/2","L/2"])
plt.savefig("Gaussian_L.png",format="png",dpi=300,bbox_inches='tight',facecolor="white",backgroundcolor="white")
```
